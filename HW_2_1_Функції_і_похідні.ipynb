1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x = sp.Symbol('x')
f = 10 * x - 5
df = sp.diff(f, x)
print(df)
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = np.full_like(x, 10)  # похідна f'(x) = 10, масив з 10 того ж розміру, що й x
```

```python
# Створення графіку
plt.figure(figsize=(10, 6))
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='red', linestyle='--')
plt.title('Функція та її похідна')
plt.xlabel('Години навчання (x)')
plt.ylabel('Оцінка / Зміна оцінки')
plt.grid(True)
plt.legend()
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
def df_calculation(f_expr):
    return sp.diff(f_expr, sp.Symbol('x'))
```

```python
def plot(f_expr, df_expr, x_range=(0, 10), num_points=1000):
    x_sym = sp.Symbol('x')

    # sympy >> numerical functions
    f_func = sp.lambdify(x_sym, f_expr, 'numpy')
    df_func = sp.lambdify(x_sym, df_expr, 'numpy')

    x_vals = np.linspace(x_range[0], x_range[1], num_points)
    y_vals = f_func(x_vals)
    dy_vals = df_func(x_vals)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, y_vals, label=f'f(x) = {f_expr}', color='blue')
    plt.plot(x_vals, dy_vals, label=f"f'(x) = {df_expr}", color='red', linestyle='--')
    plt.title('Функція та її похідна')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
x = sp.Symbol('x')
f_expr = sp.sin(x)
df = df_calculation(f_expr)

print('Похідна від функції sin(x):', df)

plot(f_expr=f_expr, df_expr=df, x_range=(0, 10))
```

```python
x = sp.Symbol('x')
f_expr_2 = sp.ln(x)
df_2 = df_calculation(f_expr_2)
print('Похідна від функції ln(x):', df_2)

plot(f_expr=f_expr_2, df_expr=df_2, x_range=(0.1, 10))
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.Symbol('x')
f_expr_3 = x**2 - 5*x + 1

df_3 = df_calculation(f_expr_3)
print('Похідна від функції f(x):', df_3)

plot(f_expr=f_expr_3, df_expr=df_3, x_range=(0, 100))
```
